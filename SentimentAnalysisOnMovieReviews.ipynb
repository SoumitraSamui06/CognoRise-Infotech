```python
import pandas as pd
import numpy as np
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
df = pd.read_csv(r"C:\Users\itsso\Downloads\Movies Dataset\IMDB Dataset.csv")
```

```python
df['review'] = df['review'].apply(word_tokenize)

stop_words = set(stopwords.words('english'))
df['review'] = df['review'].apply(lambda x: [word for word in x if word not in stop_words])

df['review'] = df['review'].apply(lambda x: [word.lower() for word in x])

df['review']=df['review'].apply(lambda x: ' '.join(x))

df['sentiment'] = df['sentiment'].map({'negative': 0, 'positive': 1})

vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(df['review'])
y = df['sentiment']
```

```python
from sklearn.naive_bayes import MultinomialNB
nb = MultinomialNB()
nb.fit(X, y)
```

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(n_estimators=100)
rf.fit(X, y)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X, y)
```

```python
import xgboost as xgb
xgb_model = xgb.XGBClassifier()
xgb_model.fit(X, y)
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
lr.fit(X, y)
```

```python
from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier()
dt.fit(X, y)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

y_pred_nb = nb.predict(X)
print("Naive Bayes Classifier:")
print("Accuracy:", accuracy_score(y, y_pred_nb))
print("Precision:", precision_score(y, y_pred_nb))
print("Recall:", recall_score(y, y_pred_nb))
print("F1 Score:", f1_score(y, y_pred_nb))

y_pred_rf = rf.predict(X)
print("Random Forest Classifier:")
print("Accuracy:", accuracy_score(y, y_pred_rf))
print("Precision:", precision_score(y, y_pred_rf))
print("Recall:", recall_score(y, y_pred_rf))
print("F1 Score:", f1_score(y, y_pred_rf))

y_pred_knn = knn.predict(X)
print("KNN Classifier:")
print("Accuracy:", accuracy_score(y, y_pred_knn))
print("Precision:", precision_score(y, y_pred_knn))
print("Recall:", recall_score(y, y_pred_knn))
print("F1 Score:", f1_score(y, y_pred_knn))

y_pred_xgb = xgb_model.predict(X)
print("XGBoosting Classifier:")
print("Accuracy:", accuracy_score(y, y_pred_xgb))
print("Precision:", precision_score(y, y_pred_xgb))
print("Recall:", recall_score(y, y_pred_xgb))
print("F1 Score:", f1_score(y, y_pred_xgb))
```

```python
y_pred_lr = lr.predict(X)
print("Logistic Regression Classifier:")
print("Accuracy:", accuracy_score(y, y_pred_lr))
print("Precision:", precision_score(y, y_pred_lr))
print("Recall:", recall_score(y, y_pred_lr))
print("F1 Score:", f1_score(y, y_pred_lr))

y_pred_dt = dt.predict(X)
print("Decision Tree Classifier:")
print("Accuracy:", accuracy_score(y, y_pred_dt))
print("Precision:", precision_score(y, y_pred_dt))
print("Recall:", recall_score(y, y_pred_dt))
print("F1 Score:", f1_score(y, y_pred_dt))
```
